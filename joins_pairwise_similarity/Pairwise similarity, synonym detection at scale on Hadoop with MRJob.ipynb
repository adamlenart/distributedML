###  Longest 5-gram (number of characters)

```python
%%writefile longest5gram.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-

import re

import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class longest5gram(MRJob):
    

    
# ------------------ settings ---------------------- #
    INPUT_PROTOCOL = RawProtocol


# ---------------- define MR steps ----------------- #

    def steps(self):
        return [MRStep(
            mapper_init=self.mapper_init,
            mapper=self.mapper,
            reducer_init = self.reducer_init,
            reducer = self.reducer,
            jobconf={
                'mapreduce.job.output.key.comparator.class': 'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
                'mapreduce.partition.keycomparator.options': '-k1,1nr',
            }
        )
    ]


# -------------- Step 1: emit records ---------------- #

    def mapper_init(self):
        '''Registers the currently longest 5 grams and their lengths'''
        self.record_lengths = 0
        self.record_5grams = []

    def mapper(self, line, freq):
        '''Looping over the whole set, emits only the currently record holding 5 grams and their lengths'''
        if len(line) >= self.record_lengths:
            self.record_lengths = len(line)
            self.record_5grams = line
            yield len(line),line


    def reducer_init(self):
        '''Counts the number of records'''
        self.n_records = 0

    def reducer(self, key, values):
        '''Emits only the first record. Expects results to be sorted by length in a decreasing order.'''
        if self.n_records == 0:
            self.n_records += 1
            for value in values:
                yield key, str(value)


    
if __name__ == '__main__':
    longest5gram.run()
```

###  Top 10 most frequent words 

```python
%%writefile mostFrequentWords.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-

import re

import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep
from collections import defaultdict

class mostFrequentWords(MRJob):


    #------------------ settings ---------------------- #
    INPUT_PROTOCOL = RawProtocol


    # ---------------- define MR steps ----------------- #

    def steps(self):
        return [
            MRStep(
                mapper=self.emit_words,
                combiner = self.emit_word_counts,
                reducer = self.emit_word_counts,
                jobconf={
                    'mapreduce.job.maps':'100',
                    'mapreduce.job.reduces':'50'
                }),
            MRStep(
                mapper = self.invert_word_count,
                reducer_init = self.define_top_n,
                reducer = self.emit_top_n,
                jobconf={
                    'mapreduce.job.output.key.comparator.class': 'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
                    'mapreduce.partition.keycomparator.options': '-k1,1nr'})
         ]
    
    # ------------------- Step 1: emit wordcounts --------- #
    
    def emit_words(self, words, freq):
        words = words.split()
        for word in words:
            yield word, 1

    def emit_word_counts(self, key, values):
        yield key, sum(values)

    def invert_word_count(self,key,value):
        yield value,key
        
    # -------------- Step 2: emit the 10 most frequent words ----- #

    def define_top_n(self):
        self.top_n = 0
        self.n = 10

    def emit_top_n(self,count,words):
        for word in words:
            if self.top_n < self.n:
                self.top_n += 1
                yield word, count
        
        
if __name__ == '__main__':
    mostFrequentWords.run()
```

### Most/Least densely appearing words 

```python
%%writefile mostLeastDenseWords.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-
from __future__ import division
import re
#import numpy as np
import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class mostLeastDenseWords(MRJob):
    
    # START STUDENT CODE 5.4.1.C

    INPUT_PROTOCOL = RawProtocol
    n = 20
    
    def steps(self):
        return [
            MRStep(
                mapper=self.emit_words_counts,
                reducer_init = self.define_word_density_containers,
                reducer = self.get_word_density,
                reducer_final = self.emit_record_densities,
                jobconf = {'mapreduce.job.maps':'150',
                           'mapreduce.job.reduces':'50'},
                ),
            MRStep(
                reducer = self.reduce_to_unique_list
            )
        ]


    def emit_words_counts(self, words, counts):
        words =  words.lower().split()
        counts = counts.split()
        n_word, n_page = counts[0],counts[1]
        for word in words:
            yield word, (n_word,n_page)

    def define_word_density_containers(self):

        self.most_dens_words = []
        self.least_dens_words = []
        self.highest_densities = []
        self.lowest_densities = []
        for n in range(self.n):
            self.most_dens_words.append("NA")
            self.least_dens_words.append("NA")
            self.highest_densities.append(0)
            self.lowest_densities.append(100000)


    def get_word_density(self, word, counts):
        n_words,n_pages = 0,0
        for n_word,n_page in counts:
            n_words += int(n_word)
            n_pages += int(n_page)

        density = n_words/n_pages
        high_index = -1
        low_index = -1
        for i in range(self.n):
            if density > self.highest_densities[i]:
                high_index = i
            elif density < self.lowest_densities[i]:
                low_index = i
            else:
                break

        if high_index >= 0:
            self.highest_densities.insert(high_index+1,density)
            self.most_dens_words.insert(high_index+1,word)
            self.highest_densities.pop(0)
            self.most_dens_words.pop(0)

        if low_index >=0:
            self.lowest_densities.insert(low_index+1,density)
            self.least_dens_words.insert(low_index+1,word)
            self.lowest_densities.pop(0)
            self.least_dens_words.pop(0)

    def emit_record_densities(self):
        # emit the most dense words
        for i in range(self.n):
            yield 'most', self.most_dens_words[i] + "," + str(self.highest_densities[i])
        # emit the least dense words
        for i in range(self.n):
            yield 'least', self.least_dens_words[i] + "," + str(self.lowest_densities[i])

    def reduce_to_unique_list(self,key,values):
        words = []
        densities = []
        for i,value in enumerate(values):
            word,density = value.split(',')
            words.append(word)
            densities.append(density)
#        # sort list of words by densities
        sorted_word = [x for (y, x) in sorted(zip(densities, words), key=lambda pair: pair[0])]
        sorted_density = sorted(densities)
        if key == 'most':
            sorted_word.reverse()
            sorted_density.reverse()
        i = 0
        for word,density in zip(sorted_word,sorted_density):
            yield key, (word,density)
            i += 1
            if i == self.n:
                break
    
    # END STUDENT CODE 5.4.1.C
        
if __name__ == '__main__':
    mostLeastDenseWords.run()
```

### Distribution of 5-gram sizes (character length)

```python
%%writefile distribution.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-

import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class distribution(MRJob):
    


    INPUT_PROTOCOL = RawProtocol

    def steps(self):
        return [
            MRStep(
                mapper = self.mapper,
                combiner = self.reducer,
                reducer = self.reducer,
                jobconf={
                    'mapreduce.job.maps':'20',
                    'mapreduce.job.reduces':'10'
                }
            )
        ]

    def mapper(self, words, freq):
        '''Emits character lengths and a count of one'''
        yield len(words),1

    def reducer(self,length,count):
        yield length,sum(count)

    
if __name__ == '__main__':
    distribution.run()
```

The Jaccard similarity is given as
\begin{align*}
J(A,B) = \frac{|A \cup B|}{|A \cap B|} = \frac{|A \cup B|}{|A| + |B| - |A \cup B|}
\end{align*}

that is, each document has length of 4,
* document A and B have 2 intersecting elements (bright, forget),
* document A and C have 2 intersecting elements (blue, bright),
* document A and D have 1 intersecting element (butterfly),
* document B and C have 2 intersectin elemtns (bright sky),
* document B and D have 1 intersecting element (sky),
* document C and D have 1 intersecting element (sky),

their Jaccard similarity is
* A-B = $\frac{2}{8-2} = \frac{1}{3}$
* A-C = $\frac{2}{8-2} = \frac{1}{3}$
* A-D = $\frac{1}{8-1} = \frac{1}{7}$

```python
%%writefile most_frequent_10k.py
import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep


class mostFrequentWords(MRJob):



# ------------------ settings ---------------------- #
    INPUT_PROTOCOL = RawProtocol
    n = 10000
    
    def remove_stopwords(self, words):
        # could, would, shall added to the list..should have added may as well
        stopwords = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'yourself',
                 'yourselves','he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them',
                 'their','theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is',
                 'are', 'was','were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an',
                 'the','and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about',
                 'against','between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up',
                 'down', 'in','out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when',
                 'where', 'why','how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor',
                 'not', 'only','own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', 'should', 'now',
                  'would','could','shall']

        return [word for word in words if word not in stopwords]

# ---------------- define MR steps ----------------- #

    def steps(self):
        return [
            MRStep(
                mapper = self.emit_words,
                combiner = self.emit_word_counts,
                reducer_init = self.define_top_n,
                reducer = self.reduce_top_n,
                reducer_final = self.emit_most_frequent,
                jobconf = {
                    'mapreduce.job.maps':'150',
                    'mapreduce.job.reduces':'10'
                }
            ),
            MRStep(
                reducer = self.reduce_to_unique_list
            )
    ]

    # -------------------------- Step 1: remove stopwords, emit most frequent words and counts-------- #


    def emit_words(self, words, freq):
        words = words.lower().split()
        words = self.remove_stopwords(words)
        for word in words:
            yield word, 1


    def emit_word_counts(self, key, values):
        yield key, sum(values)


    def define_top_n(self):
        self.most_frequent_words = []
        self.highest_counts = []
        for n in range(self.n):
            self.most_frequent_words.append("NA")
            self.highest_counts.append(0)

    def reduce_top_n(self,word,values):
        '''Calculate top N '''
        count = sum(values)
        idx = -1
        for i in range(self.n):
            if count > self.highest_counts[i]:
                idx = i
            else:
                break
        if idx >= 0:
            self.highest_counts.insert(idx+1,count)
            self.most_frequent_words.insert(idx+1,word)
            self.highest_counts.pop(0)
            self.most_frequent_words.pop(0)

    def emit_most_frequent(self):
        '''Emit top n words and counts to STDOUT'''
        for i in range(self.n):
            yield None,(self.most_frequent_words[i],self.highest_counts[i])
    
    # ---------------- Step 2: collect data from reducers and output only the top n --------- #
    def reduce_to_unique_list(self,key, values):
        '''Emit top n words collected from multiple reducers'''
        words = []
        counts = []
        for i, value in enumerate(values):
            word, count = value
            words.append(word)
            counts.append(count)
        # sort list of words by counts
        sorted_word = [x for (y, x) in sorted(zip(counts, words), key=lambda pair: pair[0])]
        sorted_word.reverse()
        i = 0
        for word in sorted_word:
            yield word, None
            # stop the loop
            i += 1
            if i == self.n: break


if __name__ == '__main__':
    mostFrequentWords.run()
```

## Step 2: prepare stripes of the most frequent 10,000 words with the most frequent 9,001-10,000 as features

```python
%%writefile stripes.py
from __future__ import division
import re
import mrjob
import json
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep
import itertools



class MRbuildStripes(MRJob):



    # ------------------ settings ---------------------- #
    INPUT_PROTOCOL = RawProtocol

    # helper function for summing stripes
    def add_dictionaries(self, dict1, dict2):
        '''Merges two dictionaries and sums up the values of the common keys'''
        new_dict = {}
        for k in itertools.chain(dict1.keys(), dict2.keys()):
            new_dict[k] = dict1.get(k, 0) + dict2.get(k, 0)
        return new_dict

    def filter_words(self,words,wordset):
         return [word for word in words if word in wordset]

    def generate_stripe_item(self,word,words,count,less_freq_words):
        # generate stripe
        #    1: get all of the other words
        other_words = words[:]
        other_words.remove(word)
        other_words = self.filter_words(other_words,less_freq_words)
        #    2: create a container for words and counts
        item = [None] * (len(other_words) * 2)
        #    3: associate the other words with their counts
        item[::2] = other_words
        item[1::2] = itertools.repeat(count, len(other_words))
        item = dict(map(None, *[iter(item)] * 2))
        #    4: place them in a dictionary to the word key
        return item




    # ---------------- define MR steps ----------------- #

    def steps(self):

        return [
            MRStep(
            mapper_init = self.mapper_init,
            mapper = self.mapper,
            reducer = self.reducer,
            jobconf = {
                    'mapreduce.job.maps':'150',
                    'mapreduce.job.reduces':'50'
                })
        ]
    # -------------- Step 1: emit stripes ---------------- #



    def mapper_init(self):
        # regex settings
        self.WORD_RE = re.compile(r"[a-z']+")
        self.COUNT_RE = re.compile(r'[0-9]+')
        #
        self.less_freq_words = []
        self.freq_words = []
        with open('most_freq_9k-10k_updated.txt','r') as infile:
            for line in infile:
                word = self.WORD_RE.search(line).group()
                self.less_freq_words.append(word)
        with open('most_freq_10k_updated.txt','r') as infile:
            for line in infile:
                word = self.WORD_RE.search(line).group()
                self.freq_words.append(word)




    def mapper(self, words, freq):
        # extract all words
        words = self.WORD_RE.findall(words.lower())
        words = self.filter_words(words,self.freq_words)
        # if there is at least a pair remaining after filtering
        if len(words) > 1:
            count = int(self.COUNT_RE.search(freq).group())
            stripe = {}
            for word in words:
                stripe[word] = self.generate_stripe_item(word,words,count,self.less_freq_words)
                if len(stripe[word]) > 0:
                    yield word, stripe[word]

    def reducer(self, key, values):
        counts = {}
        for value in values:
            counts = self.add_dictionaries(counts, value)
        yield key, counts



if __name__ == '__main__':
    MRbuildStripes.run()
```

## Step 4: calculate similarity

```python
%%writefile similarity.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-

from __future__ import division
import collections
import re
import json
import math
import numpy as np
import itertools
import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class MRsimilarity(MRJob):
  
  #START SUDENT CODE531_SIMILARITY
    
    #-------------------- define similarity functions ---------------- # 

    def get_jaccard(self,N,lenA,lenB):
        return N/(lenA+lenB-N)

    def get_cosine(self,N,lenA,lenB):
        return N/np.sqrt(lenA*lenB)
    
    def get_overlap(self,N,lenA,lenB):
        return N/min(lenA,lenB)

    def get_dice(self,N,lenA,lenB):
        return 2*N/(lenA+lenB)
    
    # ------------------------------ define MR steps ----------------- #

    def steps(self):
        JOBCONF_STEP1 = {
            'mapreduce.job.maps':'10',
            'mapreduce.job.reduces':'5'
        }
        JOBCONF_STEP2 = {
            ######### IMPORTANT: THIS WILL HAVE NO EFFECT IN -r local MODE. MUST USE -r hadoop FOR SORTING #############
            'mapreduce.job.output.key.comparator.class': 'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
            'mapreduce.partition.keycomparator.options': '-k1,1nr',
        }
        return [MRStep(jobconf=JOBCONF_STEP1,
                       mapper=self.mapper_pair_sim,
                       reducer=self.reducer_pair_sim)
            ,
                MRStep(jobconf=JOBCONF_STEP2,
                       mapper=None,
                       reducer=self.reducer_sort)
                ]

    # -------------------- Step 1: find pairs, calculate similarity  ------------------------------- #

    def mapper_pair_sim(self, _, line):
        line = line.strip()
        index, posting = line.split("\t")
        posting = json.loads(posting)

        '''
        @input: lines (postings) from inverted index
         "blue" [["DocA", 4], ["DocC", 4], ["DocE", 3]]

        @output: pairs of doc and doc_length, count the number of pairs
         make complex key and count of 1 as value:
         DocA.4.DocB.4, 1
         DocA.4.DocC.4, 1
         DocB.4.DocC.4, 1
        '''

        X = map(lambda x: x[0] + "." + str(x[1]), posting)

        # taking advantage of symetry, output only (a,b), but not (b,a)
        for subset in itertools.combinations(sorted(set(X)), 2):
            yield subset[0] + "." + subset[1], 1

    def reducer_pair_sim(self, key, value):
        '''
        @input:  
            key: Doc1_id.Doc1_len.Doc2_id.Doc2_len
            value: count
        @output: 
            key: average of similarity measures
            value: Doc1-Doc2,cosine,jaccard,overlap,dice similarity measures
        '''
        Doc1, Doc1_len, Doc2, Doc2_len = key.split(".")
        t = sum(value)
        lenDoc1,lenDoc2 = int(Doc1_len),int(Doc2_len)
        cosine = self.get_cosine(t,lenDoc1,lenDoc2)
        jaccard = self.get_jaccard(t,lenDoc1,lenDoc2)
        overlap = self.get_overlap(t,lenDoc1,lenDoc2)
        dice = self.get_dice(t,lenDoc1,lenDoc2)
        avg = np.mean([cosine,jaccard,overlap,dice])
        yield avg,(Doc1 + " - " + Doc2,cosine,jaccard,overlap,dice)

    # ----------------------------- Step 2: emit, sort ----------------------------
    def reducer_sort(self, key, value):
        for v in value:
            yield  key, v


#END SUDENT CODE531_SIMILARITY
  
if __name__ == '__main__':
    MRsimilarity.run()
```

## Evaluation of synonyms

Compare them with nltk recorded synonyms.

### Calculate performance measures:
$$Precision (P) = \frac{TP}{TP + FP} $$  
$$Recall (R) = \frac{TP}{TP + FN} $$  
$$F1 = \frac{2 * ( precision * recall )}{precision + recall}$$


We calculate Precision by counting the number of hits and dividing by the number of occurances in our top1000 (opportunities)   
We calculate Recall by counting the number of hits, and dividing by the number of synonyms in wordnet (syns)


Other diagnostic measures not implemented here:  https://en.wikipedia.org/wiki/F1_score#Diagnostic_Testing

```python
!head -n1000 similarity_updated.txt > similarity_top1000_updated.txt
```

```python
''' Performance measures '''
from __future__ import division
import numpy as np
import json
import nltk
from nltk.corpus import wordnet as wn
import sys
#print all the synset element of an element
def synonyms(string):
    syndict = {}
    for i,j in enumerate(wn.synsets(string)):
        syns = j.lemma_names()
        for syn in syns:
            syndict.setdefault(syn,1)
    return syndict.keys()
hits = []

TP = 0
FP = 0

TOTAL = 0
flag = False # so we don't double count, but at the same time don't miss hits

top1000sims = []
with open("similarity_top1000_updated.txt","r") as f:
    for line in f.readlines():

        line = line.strip()
        avg,lisst = line.split("\t")
        lisst = json.loads(lisst)
        lisst.append(avg)
        top1000sims.append(lisst)
    

measures = {}
not_in_wordnet = []

for line in top1000sims:
    TOTAL += 1

    pair = line[0]
    words = pair.split(" - ")
    
    for word in words:
        if word not in measures:
            measures[word] = {"syns":0,"opps": 0,"hits":0}
        measures[word]["opps"] += 1 
    
    syns0 = synonyms(words[0])
    measures[words[1]]["syns"] = len(syns0)
    if len(syns0) == 0:
        not_in_wordnet.append(words[0])
        
    if words[1] in syns0:
        TP += 1
        hits.append(line)
        flag = True
        measures[words[1]]["hits"] += 1
        
        
        
    syns1 = synonyms(words[1]) 
    measures[words[0]]["syns"] = len(syns1)
    if len(syns1) == 0:
        not_in_wordnet.append(words[1])

    if words[0] in syns1:
        if flag == False:
            TP += 1
            hits.append(line)
            measures[words[0]]["hits"] += 1
            
    flag = False    

precision = []
recall = []
f1 = []

for key in measures:
    p,r,f = 0,0,0
    if measures[key]["hits"] > 0 and measures[key]["syns"] > 0:
        p = measures[key]["hits"]/measures[key]["opps"]
        r = measures[key]["hits"]/measures[key]["syns"]
        f = 2 * (p*r)/(p+r)
    
    # For calculating measures, only take into account words that have synonyms in wordnet
    if measures[key]["syns"] > 0:
        precision.append(p)
        recall.append(r)
        f1.append(f)

    
# Take the mean of each measure    
print "—"*110    
print "Number of Hits:",TP, "out of top",TOTAL
print "Number of words without synonyms:",len(not_in_wordnet)
print "—"*110 
print "Precision\t", np.mean(precision)
print "Recall\t\t", np.mean(recall)
print "F1\t\t", np.mean(f1)
print "—"*110  

print "Words without synonyms:"
print "-"*100

for word in not_in_wordnet:
    print synonyms(word),word

    
```
